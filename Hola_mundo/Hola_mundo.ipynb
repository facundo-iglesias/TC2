## Introducción
*En el presente documento se procede a análizar la respuesta en frecuencia de un filtro pasabajos. Para ello, a partir de la función transferencia, se confeccionan los siguientes diagramas*:

    1. Diagramas de Bode de Magnitud y Fase.  
    2. Gráfico de polos y ceros.
    3. Esquema de Retardo de grupo.
  
### Expresión genérica para un filtro pasa-bajos:
  
\begin{equation}
H(s) = \frac{\omega_0^2}{s^2+s\frac{\omega_0}{Q}+\omega_0^2}.
\end{equation}


```python
# Librerias a utilizar
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
from scipy.signal import TransferFunction
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.figure(1)
plt.close(1)

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

```python
w0 = 0.5
qq = 1

my_tf = TransferFunction( [w0**2], [1, w0/qq, w0**2] )

bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq) )

pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq))
```
