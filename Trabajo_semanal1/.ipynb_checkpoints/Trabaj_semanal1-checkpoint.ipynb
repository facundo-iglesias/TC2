### 2) ¿Qué tipo de filtro es?

\begin{equation}
\frac{V_2}{V_1} = \frac{s- \frac{1}{R_3C_1}}{s + \frac{1}{R_3C_1}} 
\newline
s\rightarrow 0 => \frac{V_2}{V_1} = -1
\newline
s\rightarrow \infty => \frac{V_2}{V_1} = 1
\end{equation}

El filtro es de tipo pasa todo!


### 3) Obtenga la función transferencia, pero normalizada. ¿Cuál sería en este caso la norma de frecuencia y qué interpretación circuital podría tener?

\begin{equation}
\frac{V_2}{V_1} = \frac{s- \frac{1}{R_3C_1}}{s + \frac{1}{R_3C_1}} 
\end{equation}

\begin{equation}
\Omega_\omega = \frac{1}{R_3C_1}
\newline
\vec{s} = \frac{s}{\frac{1}{R_3C_1}} \rightarrow s= \frac{\vec{s}}{R_3C_1} 
\end{equation}

Función normalizada:
\begin{equation}
\frac{V_2}{V_1} = \frac{\frac{s}{R_3C_1}- \frac{1}{R_3C_1}}{\frac{s}{R_3C_1} + \frac{1}{R_3C_1}} 
\rightarrow 
\boxed
{\frac{V_2}{V_1} = \frac{s-1}{s+1}}
\end{equation}

La norma de frecuencia es:
\begin{equation}
\omega_0 = \frac{1}{R_3C_1}
\end{equation}
y representa la frecuencia a partir de la cual el filtro cambia su comportamiento invirtiendo la fase de la señal de entrada. En esta frecuencia la fase se encuentra en a 90°.




### 4) Simule la función transferencia normalizada (Python, Matlab, etc.).

```python
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
from scipy.signal import TransferFunction
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.figure(1)
plt.close(1)

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

```python

w0 = 1

my_tf = TransferFunction( [1,-w0], [1,w0] )

bodePlot(my_tf, fig_id=1, filter_description = 'w0={:3.3f}'.format(w0) )

pzmap(my_tf, fig_id=2, filter_description = 'w0={:3.3f}'.format(w0)) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3, filter_description = 'w0={:3.3f}'.format(w0))
```

### Verifique los resultados de 1 y 2 mediante el módulo de simulación simbólica SymPy.



```python
#Sympy:Libreria para resolver, expandir, simplificar y factorizar encuaciones (entre otros).
from sympy.solvers import solve
from sympy import init_session
init_session(use_latex=True)
```

```python
w = symbols('\omega',real=True)
C1 = symbols('C_1',real=True)
R3 = symbols('R_3',real=True)
P = (w*I-1/(R3*C1))
Q = (w*I+1/(R3*C1))
```

Función transferencia (s=jw)

```python
T = P/Q
T
```

Expresión del módulo:

```python
abs(T)
```

Expresión de la fase:

```python
atan(simplify(im(T)/re(T)))
```

Polos y ceros:

\begin{equation}
\frac{V_2}{V_1} = \frac{s- \frac{1}{R_3C_1}}{s + \frac{1}{R_3C_1}} 
\end{equation}

```python
P_s = (s-1/(R3*C1))
Q_s = (s+1/(R3*C1))
T_s = P_s/Q_s
```

Ceros de la función transferencia:

```python
solve(P_s, s)   
```

Polos de la función transferencia:

```python
solve(Q_s, s)
```

Análisis de los extremos de la función:

```python
from sympy import limit
lsup = limit(T_s, s, oo)
linf = limit(T_s, s, 0)
(linf,lsup)
```

\begin{equation}
Lim_{s\rightarrow0} = -1
\newline
Lim_{s\rightarrow\infty} = 1
\end{equation}
